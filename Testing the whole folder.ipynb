```python
# load model

import joblib
from matplotlib import pyplot as plt

model=joblib.load('brain_tumor_kmeans.sav')

tumorLabel=4 # brain tumor label for model
```

```python
import cv2
import os
import numpy as np

image_names=os.listdir('test_images') # get all images name
print(image_name)

# go to one by one in images
for img_name in image_names:
    
    img_path = os.path.join('test_images',img_name) # create image path
    
    imgOriginal = cv2.imread(img_path) # read image path 
    img = cv2.imread(img_path,0)
    
    height,width = img.shape # flattening
    print('img_shape :',img.shape)
    imgFlatten = img.reshape(height*width,1)
    print('imgFlatten.shape :',imgFlatten.shape)
    
    labels = model.predict(imgFlatten) # predicting
    labels2D = labels.reshape(height,width)
    
    mask = (labels2D==tumorLabel) # create mask (tumorlabel==4 -> white(1) and tumorlabel!=4 -> black(0))
    
    tumorExtracted = np.array((mask*255),dtype=np.uint8) # converting the binary array into a opencv supported image
    
    kernel = np.ones((5,5), np.uint8) # apply morphological operation -> erosion / erode
    tumorExtracted = cv2.erode(tumorExtracted, kernel, iterations=1)
    
    contours, hierarchy = cv2.findContours(tumorExtracted, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) # find contours
    print('number of contours :',len(contours))
    
    for cnt in contours:
        area = cv2.contourArea(cnt) # getting the area of the contour
        print(area)
        if(area>1000):
            cv2.drawContours(imgOriginal, [cnt], -1, (0,255,255),2) # drow contours
            x,y,w,h = cv2.boundingRect(cnt) # draw bouding rectangle
            cv2.rectangle(imgOriginal,(x,y),(x+w,y+h),(0,255,0),2) # color the bouding rectangle green
            cv2.rectangle(imgOriginal,(x,y),(x+120,y-40),(0,255,0),-1) # draw another green color rectangle
            cv2.putText(imgOriginal,"TUMOR",(x+10,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2) # writing on the rectangle
    
    cv2.imshow('LIVE',imgOriginal) # show image
    cv2.waitKey(2000)
    cv2.destroyAllWindows()
```
